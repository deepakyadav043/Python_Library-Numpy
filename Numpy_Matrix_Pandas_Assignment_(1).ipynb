## Q1. Create a 3×3 NumPy Matrix

### Input
```
1 2 3
4 5 6
7 8 9
```

### Output
```python
[[1 2 3]
 [4 5 6]
 [7 8 9]]
```


````python
import numpy as np
arr = ([[1,2,3],[4,5,6],[7,8,9]])
arr = np.array(arr)

print(arr)
````

## Q2. Multiply Matrix by 2

### Input
```python
[[1,2,3],
 [4,5,6],
 [7,8,9]]
```

### Output
```python
[[ 2  4  6]
 [ 8 10 12]
 [14 16 18]]
```


````python
import numpy as np
arr = ([[1,2,3],[4,5,6],[7,8,9]])
arr = np.array(arr)

print(arr*2)
````

## Q7. Add Bonus Column
