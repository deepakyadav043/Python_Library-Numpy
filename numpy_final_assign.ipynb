# Q1. Array Creation and Attributes

## Problem Statement
Create array [10,20,30,40,50] and print shape, ndim, size and dtype.

## Sample Input
```python
Use NumPy arrays
```

## Sample Output
```python
Expected Output
```


````python
import numpy as np
arr = [10,20,30,40,50]
arr = np.array(arr)
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
````
